```python
from sklearn import datasets
from matplotlib import pyplot as plt
from sklearn import model_selection
%matplotlib inline
```

```python
X,y = datasets.fetch_california_housing(return_X_y=True,as_frame= True)
```

```python
X
```

```python
y
```

```python
train_x,train_y,test_x,test_y = model_selection.train_test_split(X,y,test_size = 0.2)
```

```python
test_x
```

```python
train_x.plot(kind ='scatter', x = 'Longitude', y = 'Latitude',alpha = 0.4,
            s=train_x["Population"]/100, label="Population",
            c=test_x, cmap=plt.get_cmap("jet"), colorbar=True,)
plt.legend()
```
